```python
import pandas as pd 
import numpy as np
import re
```

```python
df_raw = pd.read_csv("programs.csv") 
df_raw["Season"] = df_raw["Season"].astype(str)
df_raw["Short program"] = df_raw["Short program"].astype(str)
df_raw["Free skating"] = df_raw["Free skating"].astype(str)
df_raw["Exhibition gala/ ice show"] = df_raw["Exhibition gala/ ice show"].astype(str)
```

```python
df_raw.head()
```

```python
df = pd.DataFrame(columns=["program_name"]) 
```

```python
for index, row in df_raw.iterrows():

  current_season = row["Season"]
  if index == 0:
    last_season = current_season
  if current_season != "nan":
    last_season = current_season

  if row["Short program"] != "nan":
    formatted_program = re.sub(r'\*|\"', "", row["Short program"])
    new_row = f"{last_season} {formatted_program} SP"
    print(new_row)
    df = df.append({
        "program_name": new_row
    }, ignore_index=True)

  if row["Free skating"] != "nan":
    formatted_program = re.sub(r'\*|\"', "", row["Free skating"])
    new_row = f"{last_season} {formatted_program} FS"
    df = df.append({
        "program_name": new_row
    }, ignore_index=True)

  if row["Exhibition gala/ ice show"] != "nan":
    formatted_program = re.sub(r'\*|\"', "", row["Exhibition gala/ ice show"])
    new_row = f"{last_season} {formatted_program} EX gala"
    df = df.append({
        "program_name": new_row
    }, ignore_index=True)


  
```

```python
df.head()
```

```python
df.to_csv("final.csv", index=False)
```
